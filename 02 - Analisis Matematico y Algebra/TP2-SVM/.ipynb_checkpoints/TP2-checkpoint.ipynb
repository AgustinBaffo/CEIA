## Funciones

Se definen funciones necesarias para la el calculo del gradiente descendiente.

La función `loss` calcula la función costo J, que debe minimizarse.
<img src="./img/loss.png" alt="rmse" style="width: 400px;"/>

La función `get_grad` devuelve las tres componentes del gradiente de la función costo (w1,w2,b). Las derivadas son calculadas de la siguiente forma:

<img src="./img/grad.png" alt="rmse" style="width: 500px;"/>

La función `get_line` devuelve una recta para w y b.

```python
def loss(W, xy, C=1):
    (X,Y) = xy
    X = np.array(X)
    Y = np.array(Y)
    w = np.array(W[0])
    b = np.array(W[1])
    
    acum = 0
    for x,y in zip(X,Y):
        acum += max(0,1-y*(w.T@x+b))
        
    return .5 * (np.linalg.norm(w)**2.) + C * acum

def get_grad(W, xy, C=1):
    (X,Y) = xy
    X = np.array(X)
    Y = np.array(Y)
    w = np.array(W[0])
    b = np.array(W[1])
    
    grad = np.zeros(3)
    der_b = 0.
    der_w1 = 0.
    der_w2 = 0.
    for x,y in zip(X,Y):
        if (y*(w.T@x + b) < 1):
            der_w1 += y*x[0]
            der_w2 += y*x[1]
            der_b += y
    grad[0] = w[0] - C * der_w1
    grad[1] = w[1] - C * der_w2
    grad[2] = - C * der_b
    return grad 

def get_line(x_axis,w,b):
    return -w[0]*x_axis/w[1] + b
```

## Implementación de Stochastic Gradient Descent (SGD)

En la siguiente sección se implementa el gradiente descendiente estocástico para encontrar los valores óptimos de w y b que minimizan la función costo J.

La función `SGD_update` realiza la actualización de las variables w1, w2 y b, en fución del gradiente calculado.

La función `run_SGD` realiza un loop de la cantidad de épocas pasadas como parámetro. En cada cíclo se toma una muestra random del conjunto de datos, de tamaño batch_size.

```python
def SGD_update(w, b, grad, alpha):
    w[0] = w[0] - alpha*grad[0]
    w[1] = w[1] - alpha*grad[1]
    b    = b    - alpha*grad[2]
    return w,b

def run_SGD(w_ini=[1.,1.],b_ini=0.,C=1.,learning_rate=0.001,
            batch_size=100,epoch=100,stop_error=-1):
    
    w = np.array(w_ini)
    b = np.array(b_ini)
    j_train = []
    error = 0
    first_time = True
    num_samples = samples_array.shape[0]
    
    for i in range(epoch):
        # Mezclo los datos y tomo muestra de tamaño batch_size
        shuffle(samples_array)
        batch_samples = samples_array[0:min(batch_size,num_samples)]
        x_batch = np.array(batch_samples[:,0:2])
        y_batch = np.array(batch_samples[:,2])
            
        # Obtengo el gradiente y actualizo los párametros
        grad = get_grad([w,b], xy=(x_batch,y_batch), C=C)
        w,b = SGD_update(w, b, grad, alpha = learning_rate)

        # Acumulo los errores
        j_train.append(loss([w,b], xy=(x_batch,y_batch), C=C))

        # Corte anticipado
        if(not(first_time)):            
            error = abs(j_train[-1]-j_train[-2]) # Diferencia en la función costo
            if(error < stop_error): # Si el error es menor al parametro, corto la ejecución
                print("Delta_J = "+str(error)+" < "+str(stop_error)+" => STOP")
                break
        else:
            first_time = False
    
    return w, b, j_train
```

En primer lugar, se definen los valores iniciales de w,b y los parámetros:
- learning_rate: tasa de aprendizaje
- batch_size: tamaño del lote
- epoch: cantidad de épocas
- stop_error: si ΔJ_train < stop_error, se detiene la ejecución

Luego, se computa el gradiente y se grafican los erorres de test y train.

```python
# Configuración de valores iniciales e hiperparametros
w_ini = np.random.uniform([-10,10],2)
b_ini = 10.
C = .1
learning_rate = 0.01
batch_size = 320
epoch = 300

stop_error = -1

w, b, j_train = run_SGD(w_ini, b_ini, C, learning_rate, batch_size, epoch, stop_error)
j_test = loss([w,b], xy=(X_test,y_test), C=C)
print("w = " + str(w) + ", b = "+ str(b))
print("J_train = " + str(j_train[-1]) + ", J_test = "+ str(j_test))
fig, ax = plt.subplots(figsize=[20,10])
ax.set_xlabel('Epoch', size=20)
ax.set_ylabel('Loss', size=20)
ax.plot(j_train)
```

```python
x_axis = np.linspace(-5, 5, 100)
recta = get_line(x_axis,w,b)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[20,10])
fig.suptitle(' ')
ax1.scatter(X_train[:,0], X_train[:,1], alpha=0.5, c=y_train,cmap='jet')
ax1.plot(x_axis, recta)
ax1.set_xlabel('V1')
ax1.set_ylabel('V2')
ax2.scatter(X_test[:,0], X_test[:,1], alpha=0.5, c=y_test,cmap='jet')
ax2.plot(x_axis, recta)
ax2.set_xlabel('V1')
ax2.set_ylabel('V2')
```
