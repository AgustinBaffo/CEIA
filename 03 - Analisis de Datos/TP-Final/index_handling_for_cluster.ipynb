## Explanation of index handling for cluster imputation (simple example)
https://stackoverflow.com/questions/35723472/how-to-use-sklearn-fit-transform-with-pandas-and-return-dataframe-instead-of-num

```python
import pandas as pd
import numpy as np
# Create random dataset
df = pd.DataFrame(np.random.randint(0.0,100.0,size=(10,4)),
        index=range(10,20),
        columns=['col1','col2','col3','col4'],
        dtype='float64')
        
df['name']=['alesi','spok','alesi','spok','alesi','spok','alesi','spok','alesi','spok']
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn_pandas import DataFrameMapper

# Scalce only col2 and col3 where name = 'alesi
cols = ['col2','col3']
indx=df.loc[df['name'] == 'alesi'][cols].copy().index #Get index

mapper = DataFrameMapper([(cols, StandardScaler())])
scaled_features = mapper.fit_transform(df.loc[df['name'] == 'alesi'][cols].copy(), 4)
scaled_features_df = pd.DataFrame(scaled_features, index=indx, columns=cols)
scaled_features_df
```

```python
# Replace in original dataset
df.loc[indx,cols]=scaled_features_df
df
```
