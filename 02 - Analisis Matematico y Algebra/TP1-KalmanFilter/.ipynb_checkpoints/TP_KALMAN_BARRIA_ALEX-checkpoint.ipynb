```python
def graficar(titulo, leyendas, vector, inicio=0, fin=0):
  import matplotlib.pyplot as plt

  if fin == 0:
    fin=len(vector)

  plt.figure(figsize=(10,10))
  plt.plot(vector[inicio:fin],label=leyendas)
  plt.legend()
  plt.title(titulo)

def graficar_comp(titulo, prediccion, real, medicion={}):
  plt.figure(figsize=(15,10))
  prediccion_rmse=rmse(real, prediccion)
  prediccion_mape=mape(real, prediccion)
 
  if any(medicion):
    medicion_rmse=rmse(real, medicion)
    medicion_mape=mape(real, medicion)
    plt.plot(medicion, 'r',label='Medición de '+str(titulo)+'; RMSE='+str(medicion_rmse)+'; MAPE='+str(medicion_mape))
  plt.plot(prediccion,'--',label='Predicción de '+str(titulo)+'; RMSE='+str(prediccion_rmse)+'; MAPE='+str(prediccion_mape))
  plt.plot(real,'-y',label='Valor real de '+str(titulo))
  plt.legend()
  plt.title(titulo, fontsize=20)
  return prediccion_rmse, prediccion_mape
```

```python
pos_dat=np.array(pos_dat)
vel_dat=np.array(vel_dat)
acel_dat=np.array(acel_dat)
pos_n=pos_dat+noise_pos

inicio=0
fin=350
graficar('Posición real','Posición real en el eje X',pos_n[:,0],inicio,fin)
graficar('Posición real','Posición real en el eje Y',pos_n[:,1],inicio,fin)
graficar('Posición real','Posición real en el eje Z',pos_n[:,2],inicio,fin)
```

Defino **CONDICIONES INICIALES**

```python
x_0 = np.array([10.7533, 1.1009, -5.7247, 36.6777, -17, 3.4268, -45.1769, 35.7418, 5.2774]).T 
print(x_0)
```

Defino matriz **Q**

```python
Q=0.3*np.identity(9)
print(Q)
```

Defino matriz **P**: Indica la covarianza del error en la etapa de predicción (P-) y en la etapa de corrección (P+)

```python
P=0.3*np.diag([100, 1, 0.01, 100, 1, 0.01, 100, 1, 0.01]) # Intercalo los valores para tener [x1, v1, a1, ... x3, v3, a3]
#P=0.3*np.diag([100, 100, 100, 1, 1, 1, 0.01, 0.01, 0.01])

print(P)
```

Defino la matriz **A**: es la matriz del sistema

```python
z_3=np.zeros([3,3])
A_aux=[(1, 1, 1/2), (0, 1, 1), (0, 0, 1)]
A_1=np.concatenate((A_aux,z_3,z_3))
A_2=np.concatenate((z_3,A_aux,z_3))
A_3=np.concatenate((z_3,z_3,A_aux))
A=np.concatenate((A_1,A_2,A_3),axis=1)
print(A)
```

Defino matriz **C**: Indica qué variables estoy midiendo, en este caso posición para x, y, z.

```python
c_1=[(1, 0, 0)]
c_2=[(0, 1, 0)]
z_3=[(0, 0, 0)]
C_1=np.concatenate((c_1,z_3,z_3))
C_2=np.concatenate((z_3,c_1,z_3))
C_3=np.concatenate((z_3,z_3,c_1))

C=np.concatenate((C_1,C_2,C_3),axis=1)
print(C)
```

Defino matriz **R**: Indica la covarianza del error de las mediciones, en nuestro caso el ruido.

```python
R=np.diag([100,100,100])
print(R)
```

Defino matriz B

```python
B=1*np.identity(9)
print(B)
```

Defino las mediciones que van a ingresar al filtro de Kalman. Para el punto 1 solo le paso las mediciones de posición.

```python
mediciones=pos_n
```

Defino la funcion filtroKalman a la cual se le pasan como argumentos las matrices previamente definidas y el vector de mediciones (mes)

```python
def filtroKalman(mes, P, R, C, A, Q):
  x_pred=[]
  x=x_0
  I=np.identity(9)
  

  for n in range(len(mes)):

        # prediction
    x = (A @ x)    # Predicción
    x_pred.append(x)   
    P = A @ P @ A.T + B @ Q @ B.T# Covarianza en la etapa de predicción
    

    y = mes[n] - (C@x) # Error = mediciones - C . predicción anterior

    
    K = P @ C.T @ np.linalg.inv(C @ P @ C.T + R) # Matriz de ganancias de Kalman

    x = x + (K @ y) # Predicción nueva = predicción anterior + K . (mediciones - C . predicción anterior)

    #P = (I - K @ C) @ P @ (I - K @ C).T + K@R@K.T # Covarianza en la etapa de corrección
    P = (I - K @ C) @ P

    
  return np.array(x_pred)
```

Defino las métricas por las cuales voy a evaluar el desempeño del algortimo: RMSE: Root mean squeare error
MAPE: Mean absolute percentage error

```python
def mape(real, pred):
  import numpy as np
  real, pred = np.array(real), np.array(pred)
  return np.mean(np.abs((real - pred) / real)) * 100

def rmse(real, pred):
  import numpy as np
  real, pred = np.array(real), np.array(pred)
  return (np.mean((real - pred)**2))**(0.5)
```

```python
x_pred=filtroKalman(mediciones, P, R, C, A, Q)
x_pred_gauss=filtroKalman(mediciones, P, R, C, A, Q)
error=np.array(np.zeros((9,6)))
```
